```python
import pandas as pd
import plotly.express as px
from datetime import date
```

```python
inno_geocoded = pd.read_csv("inno_geocoded.csv")
```

```python
def InnoCount(inno_nr, inno_geocoded = inno_geocoded):
    inno3 = inno_geocoded[inno_geocoded['inno_nr']==inno_nr]
    inno3["date"] = inno3['created_at'].astype('datetime64[ns]')
    inno3['Year/Week'] = inno3['date'].apply(lambda x: x.strftime("%Y/%d"))
    count = pd.DataFrame(inno3.groupby(['Year/Week', 'inno_nr', 'country_y']).size(), columns = ['count'])
    count = count.reset_index(level=0)
    return count
```

```python
count3 = InnoCount("inno_03")
```

```python
fig = px.histogram(count3, x="Year/Week", y="count", template="simple_white", title="Count per Week of Innovation 3", color_discrete_sequence=['indianred'])
fig.show()
```

```python
count31 = InnoCount("inno_31")
fig = px.histogram(count31, x="Year/Week", y="count", template="simple_white", title="Count per Week of Innovation 31", color_discrete_sequence=['darkseagreen'])
fig.show()
```

# Supercount / Per Country

```python
supercount = pd.concat([count3, count31])
supercount = supercount.reset_index(level=0)
supercount = supercount.sort_values(by=['Year/Week'])
supercount
```

```python
fig = px.histogram(supercount, x="Year/Week", y="count", template="simple_white", title="Count per Week of Innovation 03 & 31", color = "inno_nr", color_discrete_sequence=['indianred', 'darkseagreen'])
fig.show()
```

```python
supercount = pd.DataFrame(supercount.groupby(['inno_nr', 'country_y']).size(), columns = ['count'])
supercount = supercount.reset_index(level=0)
supercount = supercount.reset_index(level=0)
supercount
```

```python
fig = px.histogram(supercount, x="country_y", y="count", template="simple_white", 
    title="Count per Country of Innovation 03 & 31", color = "inno_nr", color_discrete_sequence=['indianred', 'darkseagreen'])
fig.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        title = "Country",
        tick0 = 0.5,
        dtick = 0.75
    ))
fig.show()
```
